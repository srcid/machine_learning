## Separando os dados em _features_ e _labels_

```python
X = df.iloc[:10000,:-1]
y = df['Satisfaction'].values[:10000]
```

# Preprocessamento

```python
one_hot_enc = OneHotEncoder()

features_ohe = one_hot_enc.fit_transform(X[['Gender','Customer Type','Type of Travel','Class']])
```

```python
features_labels = [lb for subl in one_hot_enc.categories_ for lb in subl]
```

```python
df_ = pd.DataFrame(features_ohe.toarray(), columns=features_labels)
```

```python
X.drop(['Gender','Customer Type','Type of Travel','Class'], axis=1, inplace=True)
```

```python
X = pd.concat([X,df_], axis=1)
```

```python
X.columns = ['Age', 'Flight Distance', 'Departure Delay', #'Arrival Delay',
       'Departure and Arrival Time Convenience', 'Ease of Online Booking',
       'Check-in Service', 'Online Boarding', 'Gate Location',
       'On-board Service', 'Seat Comfort', 'Leg Room Service', 'Cleanliness',
       'Food and Drink', 'In-flight Service', 'In-flight Wifi Service',
       'In-flight Entertainment', 'Baggage Handling', 'Female', 'Male',
       'First-time', 'Returning', 'Business Travel', 'Personal Travel', 'Business Class',
       'Economy Class', 'Economy Plus Class']
```

```python
ord_enc = OrdinalEncoder()
features_ord = ord_enc.fit_transform(y.reshape(-1,1))
```

```python
# y = pd.Series(features_ord.ravel(), name='Satisfaction')
y = features_ord.ravel()
```

# Dividindo entre treino e teste

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=80/100, random_state=42, stratify=y)
```
