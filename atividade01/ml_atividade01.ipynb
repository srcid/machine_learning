## Média, o desvio padrão, e os valores mínimo e máximo de um campo numérico

```python
dtop = df['Days In Top 10']
media = dtop.mean()
des_pad = dtop.std()
vmin = dtop.min()
vmax = dtop.max()

print('media: {}\ndesvio padrão: {}\nmin: {}\nmax: {}\n'.format(media, des_pad, vmin, vmax))
```
